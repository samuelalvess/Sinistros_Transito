# ETL: RAW ‚Üí SILVER
## Limpeza e Padroniza√ß√£o dos Dados de Sinistros de Tr√¢nsito

```python
import pandas as pd
import os

print("‚úÖ Bibliotecas carregadas")
```

```python
df_raw = pd.read_csv('../raw/datatran2025.csv', sep=';', encoding='latin-1')
print(f"Dataset original: {len(df_raw):,} registros | {len(df_raw.columns)} colunas")
```

```python
print("=" * 70)
print("LIMPEZA E PADRONIZA√á√ÉO DOS DADOS")
print("=" * 70)

registros_inicial = len(df_raw)

df_clean = df_raw.drop_duplicates()
duplicados_removidos = registros_inicial - len(df_clean)

df_clean = df_clean.replace('NA', None)

df_clean = df_clean.dropna(subset=['id', 'data_inversa', 'uf'])
registros_nulos_removidos = registros_inicial - len(df_clean)

print(f"Registros duplicados removidos: {duplicados_removidos}")
print(f"Registros com dados incompletos removidos: {registros_nulos_removidos}")
print(f"Dataset ap√≥s limpeza: {len(df_clean):,} registros v√°lidos")
```

```python
print("=" * 70)
print("PERSIST√äNCIA DOS DADOS - CAMADA SILVER")
print("=" * 70)

os.makedirs('../silver/data', exist_ok=True)

df_clean.to_csv('../silver/data/datatran2025.csv', index=False, encoding='utf-8')

print("Arquivo salvo: datatran2025.csv")
print(f"‚úÖ Transforma√ß√£o RAW ‚Üí SILVER conclu√≠da")
print(f"Total de registros: {len(df_clean):,}")
```

```python
print("=" * 70)
print("AN√ÅLISE DA TRANSFORMA√á√ÉO")
print("=" * 70)

aproveitamento = (len(df_clean) / len(df_raw) * 100)

print(f"\nüìä QUALIDADE DOS DADOS:")
print(f"  Dataset original: {len(df_raw):,} registros")
print(f"  Dataset limpo: {len(df_clean):,} registros")
print(f"  Taxa de aproveitamento: {aproveitamento:.2f}%")

print(f"\nüíæ TAMANHO:")
print(f"  Original: {df_raw.memory_usage(deep=True).sum() / 1024 / 1024:.2f} MB")
print(f"  Limpo: {df_clean.memory_usage(deep=True).sum() / 1024 / 1024:.2f} MB")
```
